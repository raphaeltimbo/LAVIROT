```python
import LaviRot as lr
from LaviRot.materials import AISI4140
%matplotlib inline
```

```python
n_el = 6
L = 1.5
s_id = 0.
s_od = 0.05

shaft = lr.ShaftElement.section(L, n_el, s_id, s_od, AISI4140)

bearing0 = lr.BearingElement(0, kxx=1e6, cxx=0)
bearing1 = lr.BearingElement(n_el, kxx=1e6, cxx=0)
bearings = [bearing0, bearing1]

disk0 = lr.DiskElement(2, AISI4140, 0.07, s_od, 0.28)
disk1 = lr.DiskElement(4, AISI4140, 0.07, s_od, 0.35)
disks = [disk0, disk1]

rotor = lr.Rotor(shaft, disks, bearings)
```

```python
lr.plot_rotor(rotor)
```

```python
speed_range = list(range(0, 480, 10))
```

```python
lr.campbell(rotor, speed_range, mult=[1, 2])
```
